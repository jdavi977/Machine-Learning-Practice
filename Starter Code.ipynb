```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load dataset
data = load_diabetes()
x = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Train model
model = LinearRegression()
model.fit(x_train, y_train)

# Evaluate model
r2 = model.score(x_test, y_test)
print('R² score:', r2)

# Examine coefficients
coefficients = pd.Series(model.coef_, index=x.columns)
print(coefficients.sort_values())

print("Question 3.2 = Based on examining the coefficients, s1 has a strong negative effect and s5 has a strong positive effect")

# Identifying the smallest coefficient
smallest_feature = coefficients.abs().idxmin()
print("Question 4.1 = The smallest coefficient feature is", smallest_feature)

# To drop a feature e.g. x_reduced = x.drop (coumns=[columnName OR variablename])

x_reduced =x.drop(columns=["age"])
x_train, x_test, y_train, y_test = train_test_split(x_reduced, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)

new_r2 = model.score(x_test, y_test)
print("Original R2 score", r2)
print('R² score:', new_r2)
print("Question 4.4 = As seen from the data, when removing the smallest coefficient age, we get a higher R2 score making a better model as it is closer to 1.")

scalar = StandardScaler()

data_scaled = scalar.fit_transform(x)
df_scaled = pd.DataFrame(data_scaled, columns=x.columns)

x_train, x_test, y_train, y_test = train_test_split(df_scaled, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

print("Original", coefficients.sort_values())

new_coefficients = pd.Series(model.coef_, index=x.columns)

print("New", new_coefficients.sort_values())

print("Question 4.a = No, the ranking of the most important features did not change")
print("Question 4.b = When scaling, the coefficient magnitudes went down significantly")
scaled_r2 = model.score(x_test, y_test)

print("Original r2:", r2)
print("Scaled r2:", scaled_r2)

print("Question 4.c = No, the R2 score did not change when scaled")

print("The features that matter the most are the feature coefficients that have the highest positive number, meaning that when the coefficent increases by 1, the targetted value will increase by that much.")
print("Yes, removing the feature improved the model slightly increasing the r2")
```
