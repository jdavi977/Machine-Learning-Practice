```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("simple_linear_regression_dataset.csv")
print(df.head())
```

```python
x_col = "x"
y_col = "y"

x = df[["x", "y"]]
y = df["z"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(mse)
print(rmse)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(mse)
print(rmse)

```

When having a 80 20 split we get a lower mse and rmse which shows that we get a closer prediction as opposed to when we use a 50 50 split. This can be seen using the mse and the rmse.
