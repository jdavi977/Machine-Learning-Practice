## Exploring Data

### Missing Data

```python
df.isnull() #show detailed view of missing data
```

```python
df.isnull().sum() #show the number of missing values per attribute
```

#### Methods for Handling Missing Data

##### 1-Drop rows/columns

```python
new_df=df.dropna(subset=['Age'], inplace=False) # remove missing data in the specified column, inplace updates the original data
```

```python
new_df.isnull().sum()
```

```python
new_df.shape
```

##### 2-Fill missing values with mean/median/mode

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.isnull().sum()
```

### Duplicate Data

```python
num_duplicates = df.duplicated().sum() #retrieves the number of duplicated rows in the dataset
print(num_duplicates)
```

```python
duplicates_name = df.duplicated(subset=['PassengerId']) #There are no duplicates
print(duplicates_name)
```

##### Handling duplicates

```python
df.drop_duplicates(inplace=True) #remving duplicates if they exist from the dataset. 
```

### Some Basic Transformations

```python
#1- Renaming Columns
df.rename(columns={'Pclass': 'Passenger_Class'}, inplace=True)
#2- Change a data type
df['Survived'] = df['Survived'].astype('category') # removes from being numerically calculated in describe()
#3 Some string operations
df['Name'] = df['Name'].str.lower() #make names lower case.
df['Cabin'] = df['Cabin'].str.strip() #remove any leading spaces
df.describe()
#df.info()
```
