### Data Preprocessing

#### Introducing the data

```python
import pandas as pd
df=pd.read_excel("Customer Call List.xlsx")
```

```python
df
```

#### Let's clean first
##### 1- Drop "Not_useful_column', drop duplicates
##### 2- Remove characters before and after 'last name'.
##### 3- Standardize 'phone number'.
##### 4- Standardize 'paying customer', and 'do_not_contact'.

```python
df.drop(columns=['Not_Useful_Column'], inplace=True)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df = df.drop(index=4)
df
```

```python
df.loc[:, 'Last_Name'] = df['Last_Name'].str.replace(r'[^a-zA-Z]', '', regex=True).str.strip()
```

```python
df
```

```python
def standardize_phone_number(phone):
    if pd.isnull(phone) or str(phone).strip() == "":
        return None
    digits = ''.join([char for char in str(phone) if char.isdigit()])
    # Format as (XXX) XXX-XXXX if it's a 10-digit number
    if len(digits) == 10:
        return f"({digits[:3]}) {digits[3:6]}-{digits[6:]}"
    else:
        return digits  # Return as-is if not 10 digits

df.loc[:, 'Phone_Number'] = df['Phone_Number'].apply(standardize_phone_number)
df.loc[:, 'Phone_Number']= df['Phone_Number'].replace(["n/a", ""], None)
df
```

```python
df.loc[:,'Paying Customer'] = df['Paying Customer'].str.strip().str.lower().replace({
        'yes': 'Yes', 'y': 'Yes', 'no': 'No', 'n': 'No'})
df.loc[:,'Paying Customer'] = df['Paying Customer'].replace("n/a", "NaN")
```

```python
df
```

```python
df.loc[:,'Do_Not_Contact'] = df['Do_Not_Contact'].str.strip().str.lower().replace({
        'yes': 'Yes', 'y': 'Yes', 'no': 'No', 'n': 'No'
    })
```

```python
df
```

```python
from datetime import datetime

df.loc[:,'Date of Birth'] = pd.to_datetime(df['Date of Birth'], errors='coerce')
df
```

```python

today = datetime.today()
df.loc[:, 'Age'] = df['Date of Birth'].apply(lambda dob: today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day)) if pd.notnull(dob) else None)
```

```python
df
```

#### Sales data Before Normalization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(df.index, df['Sale Value'], alpha=0.6, label='Original Sales')
plt.title('Scatter Plot of Original Sales Data')
plt.xlabel('Index')
plt.ylabel('Sale Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
df
```

```python
# Normalize the 'Sale Value' column using min-max normalization to be between 0-100
min_sales = df['Sale Value'].min()
max_sales = df['Sale Value'].max()
df['Normalized Sales'] = 100 * (df['Sale Value'] - min_sales) / (max_sales - min_sales)

plt.figure(figsize=(8, 5))
plt.scatter(df.index, df['Normalized Sales'], alpha=0.6, color='orange', label='Normalized Sales')
plt.title('Scatter Plot of Normalized Sales Data')
plt.xlabel('Index')
plt.ylabel('Normalized Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
age_bins = [0, 18, 35, 50, 65, 100]
age_labels = ['Child', 'Young Adult', 'Adult', 'Senior', 'Elder']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
reduced_df = df[['CustomerID', 'First_Name', 'Last_Name', 'Age', 'Age Group']]
reduced_df
```

```python
plt.figure(figsize=(8, 6))
reduced_df['Age Group'].value_counts(sort=False).plot(kind='bar')

plt.title('Histogram of Age Groups', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
stratified_sample = (
    reduced_df.groupby('Age Group', group_keys=False, observed=False)
    .apply(lambda x: x.loc[:, reduced_df.columns != 'Age Group'].sample(frac=0.5, random_state=42))
    .reset_index(drop=True)
)
stratified_sample
```
